```python
import numpy as np
import scipy as sci
import matplotlib.pyplot as pp
from mpl_toolkits import mplot3d as d3 
```

```python
A = np.array([[3,1, -4],[-1,4, 3]])  # shape (1, 2)
B = sci.linalg.orth(A.T).T

fig = pp.figure(figsize=(5,5))
dp = fig.add_subplot(111, projection = "3d")
dp.set_xlim ([-5,5])
dp.set_ylim ([-5,5])
dp.set_zlim ([-5,5])

for coord in A:
    print(coord[0],coord[1],A.shape)
    dp.quiver(0,0,0,coord[0],coord[1],coord[2], color="red")

for coord in B:
    print(coord[0], coord[1])
    dp.quiver(0,0,0,coord[0],coord[1],coord[2], color="orange")4

fig.show()
```

```python
#Ex 10-3

A = np.random.randint(0,25,36).reshape(6,6)
B = np.random.randint(0,25,36).reshape(6,6)
P,L,U = sci.linalg.lu(A)
#👇 Product of diagonal elements of a triangular matrix is equal 
# to its determinant
detU = np.prod(np.diag(U))
print(np.round(detU))
detU = np.linalg.det(U)
print(np.round(detU))

#👇 Determinant of the product matrix equals the product
# of its determinants
detA = np.linalg.det(A)
detB = np.linalg.det(B)
detAB = np.linalg.det(A@B)
print(np.round(detA),np.round( detB),np.round( detAB), np.round(detA)*np.round( detB))

# 
print(np.round(np.linalg.det(P@A)), np.round(np.prod(np.diag(U))))

```

```python
A = np.random.randint(0,25,16).reshape(4,4)
P,L,U = sci.linalg.lu(A)

# Is it a singular matrix
print("Determinant:", np.linalg.det(A))

# A = PLU
print(np.round(A))
print(np.round(P@L@U))

print("A",np.round( np.linalg.inv(A),5))
print("ULP", np.round(np.linalg.inv(U)
      @np.linalg.inv(L)
       @P,5))
```
