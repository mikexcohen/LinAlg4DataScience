```python
import numpy as np
from matplotlib import pyplot as pp
```

```python
A = np.array([13,4,12,3,14,13,12,9,11,7,13,11,9,2,5,7,10,0,9,7]).reshape(-1,1)
y  = np.array([70,25,54,21,80,68,84,62,57,40,60,64,45,38,51,52,58,21,75,70])

x = A@np.linalg.inv((A.T@A))@A.T@y
pp.scatter(A,y)
pp.plot(A,x, color='red')
pp.show()
```

```python
print(x.ravel())
print(y.ravel())
```

```python
A = np.hstack((np.ones((20,1)), np.array([13,4,12,3,14,13,12,9,11,7,13,11,9,2,5,7,10,0,9,7], ndmin=2).reshape(-1,1)))
y  = np.array([70,25,54,21,80,68,84,62,57,40,60,64,45,38,51,52,58,21,75,70])

x = A@np.linalg.inv((A.T@A))@A.T@y
E = y-x
pp.scatter(A[:,1],y)
pp.plot(A[:,1],x, color='green')
pp.show()

```

```python
print(A.shape, x.shape, y.shape)
print(x.ravel())
print(y.ravel())
print(np.dot(E,x))
print(np.corrcoef(E,x)[0,1])
```
