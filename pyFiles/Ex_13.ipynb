```python
# Eigenvector of a symmetric matrix are pair-wise orthogonal
# Symmetric matrix is a square matrix that is equal to its transpose
A = np.random.randint(-3,3,(3,3))
A = A.T@A #Symmetric matrix is created here
V,C = np.linalg.eig(A)
print(np.dot(C[:,0],C[:,1]))
print(np.dot(C[:,0],C[:,2]))
print(np.dot(C[:,1],C[:,2]))
```

```python
#An eigenvector, scaled by its norm, when np.dot'ed with itself 
#will output an identity matrix (just like any symmetric matrix)
#
N = np.linalg.norm(C, axis=0)
print(N)
CU = C/N
print(C)
print(CU)
print(np.round(CU@CU.T,2))
print (np.round(C@C.T))
```

```python
#A symmetric matrix scaled to its unit length when np.dot'ed 
#with itself will not produce an identity matrix since it is not
#orthogonal 
A = np.random.randint(-5,5,(4,4))
A = A.T@A
N = A/np.linalg.norm(A, axis=0)
print(N.T@N)

#orthogonal matrix
oN = og.rvs((3))
print(oN)
#Randome orthogonal matrix np.dot'ed with itself creates 
#an identity matrix 
print(np.round(oN.T@oN,2))
```
